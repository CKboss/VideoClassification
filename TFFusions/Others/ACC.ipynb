```python
# load label
Labels = dict()
with open('lsvc_class_index.txt','r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.replace('\n','')
        line = line.replace(' ','')
        ID,Name = line.split('\t')
        Labels[int(ID)-1] = Name
```
